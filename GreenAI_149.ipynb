```python
#write a code for conditional statements in Python
# This function adds two numbers
def add(x, y):
    return x + y

# This function subtracts two numbers
def subtract(x, y):
    return x - y

# This function multiplies two numbers
def multiply(x, y):
    return x * y

# This function divides two numbers
def divide(x, y):
    return x / y


print("Select operation.")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")

while True:
    # take input from the user
    choice = input("Enter choice(1/2/3/4): ")

    # check if choice is one of the four options
    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        if choice == '1':
            print(num1, "+", num2, "=", add(num1, num2))

        elif choice == '2':
            print(num1, "-", num2, "=", subtract(num1, num2))

        elif choice == '3':
            print(num1, "*", num2, "=", multiply(num1, num2))

        elif choice == '4':
            print(num1, "/", num2, "=", divide(num1, num2))

        # check if user wants another calculation
        # break the while loop if answer is no
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation == "no":
          break
    else:
        print("Invalid Input")
```

```python
def main():
    # Initialize a list to store temperatures
    temperatures = []
    city_count = 5

    # Prompt user for yearly temperatures of five cities
    for i in range(city_count):
        temp = float(input(f"Enter the yearly temperature for city {i + 1}: "))
        temperatures.append(temp)

    # Calculate maximum and minimum temperatures
    max_temp = max(temperatures)
    min_temp = min(temperatures)

    # Display the results
    print(f"The maximum temperature recorded is: {max_temp}째C")
    print(f"The minimum temperature recorded is: {min_temp}째C")

# Call the main function
if __name__ == "__main__":
    main()
```

```python
def calculate_carbon_footprint(electricity_usage_kwh, car_mileage_miles, waste_kg):
    # Constants for CO2 emissions (in kg)
    electricity_factor = 0.5  # kg CO2 per kWh
    car_factor = 0.404  # kg CO2 per mile driven
    waste_factor = 0.8  # kg CO2 per kg of waste

    # Calculate the carbon footprint
    electricity_footprint = electricity_usage_kwh * electricity_factor
    car_footprint = car_mileage_miles * car_factor
    waste_footprint = waste_kg * waste_factor

    total_footprint = electricity_footprint + car_footprint + waste_footprint

    return total_footprint

# Example usage
if __name__ == "__main__":
    # Input data
    electricity_usage = float(input("Enter electricity usage in kWh: "))
    car_mileage = float(input("Enter car mileage in miles: "))
    waste_generated = float(input("Enter waste generated in kg: "))

    carbon_footprint = calculate_carbon_footprint(electricity_usage, car_mileage, waste_generated)
    print(f"Your annual carbon footprint is approximately: {carbon_footprint:.2f} kg CO2.")
```

```python
#use lambda function to check given year is leap year or not
is_leap_year = lambda year: year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

# Get the year from the user
year = int(input("Enter a year: "))

# Check if it's a leap year using the lambda function
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#to check weather a strring is palindeome or not using numpy
import numpy as np

def is_palindrome(s):
    # Remove spaces and convert to lowercase for uniformity
    s = s.replace(" ", "").lower()
    # Convert the string into a numpy array
    arr = np.array(list(s))
    # Check if the array is equal to its reverse
    return np.array_equal(arr, arr[::-1])

# Example usage
string = "madam"
if is_palindrome(string):
    print(f"'{string}' is a palindrome.")
else:
    print(f"'{string}' is not a palindrome.")
```
